# Task 4: Web Scraping Product Data

This program extracts product name, price, and rating from an e-commerce website and stores it in a CSV file.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Website URL
url = "https://books.toscrape.com/"

# Send request
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Lists to store data
names = []
prices = []
ratings = []

# Find all products
products = soup.find_all("article", class_="product_pod")

for product in products:
    # Name
    name = product.h3.a["title"]
    names.append(name)

    # Price
    price = product.find("p", class_="price_color").text
    prices.append(price)

    # Rating
    rating_class = product.find("p")["class"][1]
    ratings.append(rating_class)

# Create DataFrame
data = pd.DataFrame({
    "Product Name": names,
    "Price": prices,
    "Rating": ratings
})

# Save to CSV
data.to_csv("products.csv", index=False)

print("✅ Data saved to products.csv")
print(data.head())
```

## Explanation

- The program uses requests to fetch webpage data.
- BeautifulSoup parses the HTML content.
- It extracts product name, price, and rating.
- The data is stored in a pandas DataFrame.
- Finally, it is saved as a CSV file.
